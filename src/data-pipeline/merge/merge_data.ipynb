```python
cols = [
    'date_mutation',
    'valeur_fonciere',
    'surface_reelle_bati',
    'code_commune',
    'type_local',
    'nombre_pieces_principales',
    'longitude',
    'latitude',
    'code_postal',
    'id_mutation'
]

import os
import pandas as pd

def prepare_and_agg(df):

    os.makedirs('raw', exist_ok=True)
    
    df = df[cols].copy()
    df['prix_m2'] = df['valeur_fonciere'] / df['surface_reelle_bati']
    df['annee'] = pd.to_datetime(df['date_mutation']).dt.year
    
    df_agg = df.groupby(['code_commune', 'annee']).agg({
        'prix_m2': 'mean',
        'surface_reelle_bati': 'mean',
        'id_mutation': 'count'
    }).rename(columns={'id_mutation':'nb_ventes'}).reset_index()
    
    annee= df['annee'].unique()[0] 
    output_path = os.path.join('raw', f'csv_agg_{annee}.csv')
    
    df_agg.to_csv(output_path, index=False, sep=';', encoding='utf-8')
    
    print(f"Fichier CSV enregistré ici : {output_path}")
    return df_agg
```

```python
vf_annee = df.groupby(['annee', 'code_commune']).agg(
    prix_m2=('prix_m2', 'mean'),
    nom_commune=('nom_standard', 'first'),
    dep_code=('dep_code', 'first'),
    dep_nom=('dep_nom', 'first'),
    reg_code=('reg_code', 'first'),
    reg_nom=('reg_nom', 'first'),
    code_postal=('code_postal', 'first'),
    population=('population', 'first'),
    densite=('densite', 'first'),
    superficie_km2=('superficie_km2', 'first'),
    zone_emploi=('zone_emploi', 'first'),
    taux_global_tfb=('taux_global_tfb', 'first'),
    taux_global_tfnb=('taux_global_tfnb', 'first'),
    taux_plein_teom=('taux_plein_teom', 'first'),
    taux_global_th=('taux_global_th', 'first'),
).reset_index()


def market_trend(x):
    if x > 0:
        return 'hausse'
    elif x < 0:
        return 'baisse'
    else:
        return 'stable'
    
vf_annee['variation'] = vf_annee.groupby('code_commune')['prix_m2'].diff()
vf_annee['y'] = vf_annee['variation'].apply(market_trend)
```
