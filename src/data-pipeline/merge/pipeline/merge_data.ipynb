```python
cols = [
    'date_mutation',
    'valeur_fonciere',
    'surface_reelle_bati',
    'code_commune',
    'type_local',
    'nombre_pieces_principales',
    'longitude',
    'latitude',
    'code_postal',
    'id_mutation'
]

import os
import pandas as pd

def prepare_and_agg(df):

    os.makedirs('raw', exist_ok=True)
    
    df = df[cols].copy()
    df['prix_m2'] = df['valeur_fonciere'] / df['surface_reelle_bati']
    df['annee'] = pd.to_datetime(df['date_mutation']).dt.year
    
    df_agg = df.groupby(['code_commune', 'annee']).agg({
        'prix_m2': 'mean',
        'surface_reelle_bati': 'mean',
        'id_mutation': 'count'
    }).rename(columns={'id_mutation':'nb_ventes'}).reset_index()
    
    annee= df['annee'].unique()
    output_path = os.path.join('raw', f'csv_agg_{annee}.csv')
    
    # Sauvegarde en CSV
    df_agg.to_csv(output_path, index=False, sep=';', encoding='utf-8')
    
    print(f"Fichier CSV enregistré ici : {output_path}")
    
    return df_agg
```

```python
df['date_mutation'] = pd.to_datetime(df['date_mutation'])

annual_cp = df.groupby([df['code_postal'], df['date_mutation'].dt.to_period('Y')]).agg({
    'valeur_fonciere': 'mean'
}).sort_index()

annual_cp['variation'] = annual_cp.groupby(level=0)['valeur_fonciere'].diff()

def trend(x):
    if x > 0:
        return 'hausse'
    if x < 0:
        return 'baisse'
    return 'stable'

annual_cp['tendance'] = annual_cp['variation'].fillna(0).apply(trend)

df['année'] = df['date_mutation'].dt.year
df['y'] = df.apply(lambda row: 
                   annual_cp.loc[(row['code_postal'], pd.Period(row['date_mutation'], freq='Y')), 'tendance'], 
                   axis=1)

df.head()

```
