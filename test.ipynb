```python
import pandas as pd
```

```python
df = pd.read_csv('full.csv')
df.head()
```

```python
df['date_mutation'] = pd.to_datetime(df['date_mutation'])
df
```

```python
pd.options.display.float_format = '{:,.0f}'.format
```

```python
monthly = df.groupby(df['date_mutation'].dt.to_period('M'), as_index=False).agg({'valeur_fonciere': 'mean'})
monthly
```

```python
monthly_change = monthly.set_index('date_mutation')['valeur_fonciere'].diff()

def trend(x):
    if x > 0:
        return 'hausse'
    if x < 0:
        return 'baisse'
    return 'stable'

trend_map = monthly_change.fillna(0).apply(trend)
df['y'] = df['date_mutation'].dt.to_period('M').map(trend_map)
```
